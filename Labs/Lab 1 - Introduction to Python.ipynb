#### Task
Write a function (gisc_exponential) to calculate the exponential of a variable with the base of Eulerâ€™s number. Return 0 if the result is smaller than 0.1.

```python
def gisc_exponential(number):
    """Return exponential of number, or zero if results is smaller than 0.1"""
    from numpy import exp
    result = exp(number)
    if result < 0.1:
        result = 0
    return result
```
